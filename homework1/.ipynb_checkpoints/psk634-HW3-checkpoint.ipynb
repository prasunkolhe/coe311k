```python
import numpy as np
import matplotlib.pyplot as plt
import math

def taylor_sin(x, n_terms):
    """
    Truncated Taylor series for sin(x) about x = 0:
    sin(x)  _{k=0}^{n_terms-1} (-1)^k * x^(2k+1) / (2k+1)!
    Works for x as a number or a numpy array.
    """
    s = 0.0
    for k in range(n_terms):
        power = 2 * k + 1
        term = ((-1) ** k) * (x ** power) / math.factorial(power)
        s = s + term
    return s

# x values from 0 to 10 with step 0.1
x = np.arange(0, 10.0 + 0.1, 0.1)

# exact values using numpy
sin_exact = np.sin(x)

# compute approximations and absolute errors for 1, 2, 3, 4 terms
terms_list = [1, 2, 3, 4]
sin_approx = {}
abs_error = {}

for n in terms_list:
    sin_approx[n] = taylor_sin(x, n)
    abs_error[n] = np.abs(sin_exact - sin_approx[n])

#sample values so the output is visible in the notebook
sample_x = [0.0, 0.5, 1.0, 2.0, 5.0, 10.0]
print("Sample results (sin exact vs Taylor approx)".center(70, "-"))
print(f"{'x':>6} {'sin(x)':>12} {'1 term':>12} {'2 terms':>12} {'3 terms':>12} {'4 terms':>12}")
for xv in sample_x:
    idx = int(round(xv / 0.1))
    print(f"{x[idx]:6.1f} {sin_exact[idx]:12.6f} {sin_approx[1][idx]:12.6f} {sin_approx[2][idx]:12.6f} "
          f"{sin_approx[3][idx]:12.6f} {sin_approx[4][idx]:12.6f}")

print("\nAbsolute error at x = 10".center(70, "-"))
idx10 = int(round(10.0 / 0.1))
for n in terms_list:
    print(f"{n} term(s): |error| = {abs_error[n][idx10]:.6e}")

# plot truncation error vs x for 1, 2, 3, 4 terms
plt.figure()
for n in terms_list:
    plt.plot(x, abs_error[n], label=f"{n} term(s)")
plt.xlabel("x")
plt.ylabel("Absolute error |sin(x) - Taylor(x)|")
plt.title("Truncation Error for sin(x) Taylor Series (about 0)")
plt.legend()
plt.grid(True)
plt.show()

# short discussion
print("Discussion")
print(
    "As you include more Taylor series terms, the truncation error generally decreases.\n"
    "But the error does not decrease uniformly for all x values: it stays small near x = 0 and\n"
    "tends to grow as x gets larger because the Taylor series here is centered at 0.\n"
    "So for larger x, you usually need more terms to keep the approximation accurate."
)
```
