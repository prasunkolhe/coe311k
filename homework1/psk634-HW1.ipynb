```python
import random
import math
import matplotlib.pyplot as plt

# Note:
# Citation: Parts of this code (structure, debugging guidance, and plotting approach) were developed with assistance from ChatGPT (OpenAI), Feb 2026.


def montecarlo_pi(number_of_shots):
    """
    Estimate pi using the Monte Carlo method.
    """
    number_of_hits = 0

    for _ in range(number_of_shots):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)

        if x*x + y*y < 1:
            number_of_hits += 1

    return 4 * number_of_hits / number_of_shots


# Main code
random.seed(311)

pi_true = math.pi

samples = list(range(10, 1001, 20))
errors = []

for n in samples:
    pi_est = montecarlo_pi(n)
    error = abs(pi_est - pi_true)
    errors.append(error)

# Print pi to 15 digits past the decimal
print(f"Actual pi (15 digits) = {pi_true:.15f}")

# Plot error vs number of samples
plt.figure()
plt.plot(samples, errors, marker='o')
plt.xlabel("Number of samples")
plt.ylabel("Absolute error")
plt.title("Monte Carlo Pi Approximation Error")
plt.grid(True)
plt.show()
```
