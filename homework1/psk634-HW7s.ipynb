# Stability Analysis for dy/dx = -2y cos(x)

## Given ODE
\[
\frac{dy}{dx} = -2y\cos(x)
\]

This is a linear, time-varying ODE of the form:
\[
\frac{dy}{dx} = \lambda(x)\,y
\]
where:
\[
\lambda(x) = -2\cos(x)
\]

---

## Exact Solution (for stability insight)

Separate variables:
\[
\frac{1}{y}dy = -2\cos(x)\,dx
\]

Integrate:
\[
\ln|y| = -2\sin(x) + C
\]

Exponentiate:
\[
y(x) = Ce^{-2\sin(x)}
\]

Apply initial condition \(y(0)=1\):
\[
y(x) = e^{-2\sin(x)}
\]

---

## Euler Method Update

Euler’s method:
\[
y_{n+1} = y_n + h f(x_n, y_n)
\]

Substitute \(f(x,y) = -2y\cos(x)\):
\[
y_{n+1} = y_n (1 - 2h\cos(x_n))
\]

---

## Stability Condition

For stability:
\[
|1 - 2h\cos(x)| < 1
\]

Worst case occurs when:
\[
|\cos(x)| = 1
\]

So:
\[
|1 \pm 2h| < 1
\]

Solve:
\[
-1 < 1 - 2h < 1
\]

From left inequality:
\[
-1 < 1 - 2h \Rightarrow h < 1
\]

From right inequality:
\[
1 - 2h < 1 \Rightarrow h > 0
\]

---

## Final Stability Condition

\[
\boxed{0 < h < 1}
\]

---

## Conclusion

The stability of Euler’s method depends on both the step size \(h\) and the variation of \(\cos(x)\). Since \(\cos(x)\) ranges between -1 and 1, the worst-case scenario determines stability. Euler’s method remains stable only when \(0 < h < 1\). For larger step sizes, the numerical solution becomes unstable and may diverge.
