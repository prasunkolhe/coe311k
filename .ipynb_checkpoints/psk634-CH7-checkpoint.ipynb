```python
import numpy as np

# ChatGPT helped me add max_iter + a divergence check.

def jacobi(A, b, tol=1e-6, max_iter=5000, x0=None):
    A = np.array(A, float)
    b = np.array(b, float)
    n = len(b)

    x = np.zeros(n) if x0 is None else np.array(x0, float)
    x_new = np.zeros(n)

    for it in range(1, max_iter + 1):
        with np.errstate(over="ignore", invalid="ignore"):
            for i in range(n):
                s = 0.0
                for j in range(n):
                    if j != i:
                        s += A[i, j] * x[j]
                x_new[i] = (b[i] - s) / A[i, i]

        if not np.all(np.isfinite(x_new)):
            return x_new, it, False

        if np.max(np.abs(x_new - x)) < tol:
            return x_new, it, True

        x = x_new.copy()

    return x, max_iter, False


def gauss_seidel(A, b, tol=1e-6, max_iter=5000, x0=None):
    A = np.array(A, float)
    b = np.array(b, float)
    n = len(b)

    x = np.zeros(n) if x0 is None else np.array(x0, float)

    for it in range(1, max_iter + 1):
        x_old = x.copy()

        with np.errstate(over="ignore", invalid="ignore"):
            for i in range(n):
                s = 0.0
                for j in range(n):
                    if j != i:
                        s += A[i, j] * x[j]
                x[i] = (b[i] - s) / A[i, i]

        if not np.all(np.isfinite(x)):
            return x, it, False

        if np.max(np.abs(x - x_old)) < tol:
            return x, it, True

    return x, max_iter, False


def run_case(name, A, b, x0=None):
    xj, itj, okj = jacobi(A, b, x0=x0)
    xg, itg, okg = gauss_seidel(A, b, x0=x0)
    print(name)
    print("Jacobi:", okj, "iters:", itj)
    print("Gauss-Seidel:", okg, "iters:", itg)
    print("x:", xg)
    print()


A1 = [
    [0.582745, 0.48,     0.10,    0.0,     0.0],
    [0.48,     1.044129, 0.46,    0.10,    0.0],
    [0.10,     0.46,     1.10431, 0.44,    0.10],
    [0.0,      0.10,     0.44,    0.963889,0.42],
    [0.0,      0.0,      0.10,    0.42,    0.522565]
]
b1 = [1.162745, 2.084129, 2.20431, 1.923889, 1.042565]

A2 = [[1, 1],
      [1, 1.0001]]

b2 = [1, 1]
b3 = [1, 1.0001]

run_case("Case 1", A1, b1)                 
run_case("Case 2", A2, b2)                
run_case("Case 3", A2, b3, x0=[0, 1])     
```
