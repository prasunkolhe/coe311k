```python
n = 2025
remainders = []

print(f"{'Division':<12}{'Quotient':<12}{'Remainder'}")

while n > 0:
    quotient = n // 2
    remainder = n % 2

    division_str = f"{n} ÷ 2"
    print(f"{division_str:<12}{quotient:<12}{remainder}")

    remainders.append(remainder)
    n = quotient

binary = ''.join(str(bit) for bit in reversed(remainders))
print("\nBinary =", binary)
```
