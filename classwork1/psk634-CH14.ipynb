```python
import numpy as np
import matplotlib.pyplot as plt

# ChatGPT was used to help organize and check this code.

a = 0
b = 2 * np.pi
N = 100

x = np.linspace(a, b, N)
h = (b - a) / (N - 1)

f = np.sin(x)
true_derivative = np.cos(x)

# Central Difference Approximation
central = (np.sin(x + h) - np.sin(x - h)) / (2 * h)

# Forward Difference Approximation
forward = (np.sin(x + h) - np.sin(x)) / h

# Backward Difference Approximation
backward = (np.sin(x) - np.sin(x - h)) / h

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, true_derivative)
axs[0, 0].set_title("True Derivative: cos(x)")
axs[0, 0].grid(True)

axs[0, 1].plot(x, forward)
axs[0, 1].set_title("Forward Difference")
axs[0, 1].grid(True)

axs[1, 0].plot(x, backward)
axs[1, 0].set_title("Backward Difference")
axs[1, 0].grid(True)

axs[1, 1].plot(x, central)
axs[1, 1].set_title("Central Difference")
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
