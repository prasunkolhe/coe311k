```python
import numpy as np
import matplotlib.pyplot as plt

# ChatGPT was used to help confirm the least-squares setup

# ---------- Part 1 ----------
X = np.array([1,2,3,4,5], dtype=float)
Y = np.array([1.2,2.8,3.6,4.5,5.1], dtype=float)

A = np.vstack([X, np.ones(len(X))]).T
m1, c1 = np.linalg.lstsq(A, Y, rcond=None)[0]
yfit1 = m1*X + c1

print("Part 1")
print(f"Slope: {m1:.4f}, Intercept: {c1:.4f}")

plt.figure()
plt.scatter(X, Y)
plt.plot(X, yfit1)
plt.title("Part 1")
plt.show()

# ---------- Part 2 ----------
rng = np.random.default_rng(311)
Y_rand = rng.permutation(Y)

m2, c2 = np.linalg.lstsq(A, Y_rand, rcond=None)[0]
yfit2 = m2*X + c2

print("\nPart 2")
print(f"Slope: {m2:.4f}, Intercept: {c2:.4f}")

plt.figure()
plt.scatter(X, Y_rand)
plt.plot(X, yfit2)
plt.title("Part 2 (Randomized)")
plt.show()

# ---------- Part 3 ----------
m_true, c_true = 2.0, 1.0
x3 = np.arange(0,1.01,0.01)
y3 = m_true*x3 + c_true + rng.uniform(-0.1,0.1,len(x3))

A3 = np.vstack([x3, np.ones(len(x3))]).T
m3, c3 = np.linalg.lstsq(A3, y3, rcond=None)[0]
yfit3 = m3*x3 + c3

print("\nPart 3")
print(f"True line: y = {m_true:.2f}x + {c_true:.2f}")
print(f"Fit line:  y = {m3:.2f}x + {c3:.2f}")

plt.figure()
plt.scatter(x3, y3, s=10)
plt.plot(x3, yfit3)
plt.title("Part 3")
plt.show()
```
