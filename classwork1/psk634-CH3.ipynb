```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Note:
# Code structure and debugging assistance were supported by ChatGPT (OpenAI), Feb 2026.

def montecarlo_pi(number_of_shots):
    number_of_hits = 0

    for _ in range(number_of_shots):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)

        if x*x + y*y < 1:
            number_of_hits += 1

    return 4 * number_of_hits / number_of_shots


# Main code
random.seed(311)

pi_true = np.pi
num_runs = 100
samples_per_run = 1000

absolute_errors = []
relative_errors = []

for _ in range(num_runs):
    pi_est = montecarlo_pi(samples_per_run)

    abs_error = abs(pi_true - pi_est)
    rel_error = abs(pi_true - pi_est) / pi_true

    absolute_errors.append(abs_error)
    relative_errors.append(rel_error)


# Plot Absolute Error
plt.figure()
plt.plot(range(1, num_runs + 1), absolute_errors, marker='o')
plt.xlabel("Run number")
plt.ylabel("Absolute Error")
plt.title("Absolute Error of Monte Carlo Pi (100 runs, 1000 samples)")
plt.grid(True)
plt.show()

# Plot Relative Error
plt.figure()
plt.plot(range(1, num_runs + 1), relative_errors, marker='o')
plt.xlabel("Run number")
plt.ylabel("Relative Error")
plt.title("Relative Error of Monte Carlo Pi (100 runs, 1000 samples)")
plt.grid(True)
plt.show()
```
