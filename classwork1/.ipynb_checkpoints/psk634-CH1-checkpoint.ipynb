```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)

hist_data = np.random.normal(0, 1, 500)

bar_labels = ['A', 'B', 'C', 'D']
bar_values = [5, 9, 3, 7]

plt.figure(figsize=(10, 8))

# 1) Line plot
plt.subplot(2, 2, 1)
plt.plot(x, y, 'b--', linewidth=2)
plt.title("Line Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)

# 2) Scatter plot
plt.subplot(2, 2, 2)
plt.scatter(scatter_x, scatter_y, color='red', marker='x')
plt.title("Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# 3) Histogram
plt.subplot(2, 2, 3)
plt.hist(hist_data, bins=20, color='green', edgecolor='black')
plt.title("Histogram")
plt.xlabel("value")
plt.ylabel("frequency")

# 4) Bar chart (other graph)
plt.subplot(2, 2, 4)
plt.bar(bar_labels, bar_values, color='purple')
plt.title("Bar Chart")
plt.xlabel("category")
plt.ylabel("value")

plt.tight_layout()
plt.show()
```
