```python
import numpy as np

# Note:
# Some guidance on structuring this loop and computing relative error
# was assisted by ChatGPT (OpenAI), Feb 2026.


# Leibniz series for pi:
# pi = 4 * (1 - 1/3 + 1/5 - 1/7 + ...)

pi_true = np.pi
partial_sum = 0.0

num_terms = 5000

for n in range(num_terms):
    term = (-1)**n / (2*n + 1)
    partial_sum += term

    pi_est = 4 * partial_sum

    # Every 100 terms, print relative error
    if (n + 1) % 100 == 0:
        relative_error = abs(pi_est - pi_true) / pi_true
        print(f"Terms: {n+1:4d} | Relative Error: {relative_error:.6e}")
```
