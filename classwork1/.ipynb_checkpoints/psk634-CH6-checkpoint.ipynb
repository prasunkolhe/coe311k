```python
import numpy as np

def jacobi(A, b, tol=1e-6, max_iter=5000):
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)

    x = np.zeros(n)
    x_new = np.zeros(n)

    for it in range(1, max_iter + 1):
        for i in range(n):
            s = 0.0
            for j in range(n):
                if j != i:
                    s += A[i][j] * x[j]
            x_new[i] = (b[i] - s) / A[i][i]

        if np.max(np.abs(x_new - x)) < tol:
            return x_new, it, True

        x = x_new.copy()

    return x, max_iter, False


def gauss_seidel(A, b, tol=1e-6, max_iter=5000):
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)

    x = np.zeros(n)

    for it in range(1, max_iter + 1):
        x_old = x.copy()

        for i in range(n):
            s = 0.0
            for j in range(n):
                if j != i:
                    s += A[i][j] * x[j]
            x[i] = (b[i] - s) / A[i][i]

        if np.max(np.abs(x - x_old)) < tol:
            return x, it, True

    return x, max_iter, False


A6a = [
    [10, -1,  2,  0,  0],
    [-1, 11, -1,  3,  0],
    [ 2, -1, 10, -1,  0],
    [ 0,  3, -1,  8, -2],
    [ 0,  0,  0, -2,  9]
]
b6a = [14, 30, 26, 25, 37]

xj, itj, okj = jacobi(A6a, b6a)
xg, itg, okg = gauss_seidel(A6a, b6a)

print("6a")
print("Jacobi converged:", okj, "iters:", itj)
print("Gauss-Seidel converged:", okg, "iters:", itg)
print("x:", xg)
```

```python
import numpy as np

def jacobi(A, b, tol=1e-6, max_iter=5000):
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)

    x = np.zeros(n)
    x_new = np.zeros(n)

    for it in range(1, max_iter + 1):
        with np.errstate(over="ignore", invalid="ignore"):
            for i in range(n):
                s = 0.0
                for j in range(n):
                    if j != i:
                        s += A[i][j] * x[j]
                x_new[i] = (b[i] - s) / A[i][i]

        if not np.all(np.isfinite(x_new)):
            return x_new, it, False

        if np.max(np.abs(x_new - x)) < tol:
            return x_new, it, True

        x = x_new.copy()

    return x, max_iter, False


def gauss_seidel(A, b, tol=1e-6, max_iter=5000):
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)

    x = np.zeros(n)

    for it in range(1, max_iter + 1):
        x_old = x.copy()

        with np.errstate(over="ignore", invalid="ignore"):
            for i in range(n):
                s = 0.0
                for j in range(n):
                    if j != i:
                        s += A[i][j] * x[j]
                x[i] = (b[i] - s) / A[i][i]

        if not np.all(np.isfinite(x)):
            return x, it, False

        if np.max(np.abs(x - x_old)) < tol:
            return x, it, True

    return x, max_iter, False


# 6b system
A6b = [
    [1, 2, 3, 0, 0],
    [2, 1, 2, 3, 0],
    [3, 2, 1, 2, 3],
    [0, 3, 2, 1, 2],
    [0, 0, 3, 2, 1]
]
b6b = [14, 22, 33, 26, 22]

xj, itj, okj = jacobi(A6b, b6b)
xg, itg, okg = gauss_seidel(A6b, b6b)

print("6b")
print("Jacobi converged:", okj, "iters:", itj)
print("Gauss-Seidel converged:", okg, "iters:", itg)
print("x last:", xg)
```
