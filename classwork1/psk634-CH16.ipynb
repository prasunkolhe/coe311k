```python
import numpy as np
import matplotlib.pyplot as plt

# ChatGPT was used to help organize and check this code.

def euler_oscillator(h, k, x_start, x_end, y0, v0):
    x_values = np.arange(x_start, x_end + h, h)
    if x_values[-1] > x_end:
        x_values = x_values[:-1]

    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    y_values[0] = y0
    v_values[0] = v0

    for i in range(1, len(x_values)):
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]

        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    return x_values, y_values

# Parameters
k = 1.0
x_start = 0
x_end = 10
y0 = 1.0
v0 = 0.0

h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]

plt.figure(figsize=(10, 6))

for h in h_values:
    x_vals, y_vals = euler_oscillator(h, k, x_start, x_end, y0, v0)
    plt.plot(x_vals, y_vals, label=f"h = {h}")

plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Euler Method Stability for Harmonic Oscillator (k=1)")
plt.legend()
plt.grid(True)
plt.show()

# Test instability with larger h
large_h_values = [1.0, 1.5, 2.0]

plt.figure(figsize=(10, 6))

for h in large_h_values:
    x_vals, y_vals = euler_oscillator(h, k, x_start, x_end, y0, v0)
    plt.plot(x_vals, y_vals, label=f"h = {h}")

plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Instability for Larger Step Sizes")
plt.legend()
plt.grid(True)
plt.show()

# Change k value
k = 10

plt.figure(figsize=(10, 6))

for h in [0.01, 0.05, 0.1]:
    x_vals, y_vals = euler_oscillator(h, k, x_start, x_end, y0, v0)
    plt.plot(x_vals, y_vals, label=f"h = {h}")

plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Effect of Larger k (k=10)")
plt.legend()
plt.grid(True)
plt.show()
```

Euler’s method is stable for small step sizes because it closely follows the true oscillatory behavior. As the step size increases, the solution begins to grow instead of oscillating properly, showing instability. This happens because Euler’s method does not conserve energy in oscillatory systems, causing the amplitude to increase over time. For larger values of k, the system oscillates faster, so a smaller step size is required to maintain stability. When h is too large, the solution diverges quickly, showing that Euler’s method is not suitable for solving oscillatory problems with large step sizes.
