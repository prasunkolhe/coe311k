```python
import numpy as np
import matplotlib.pyplot as plt

# ChatGPT was used to help organize and check this code.

k = 1.0
x_start = 0
x_end = 10
y0 = 1.0
v0 = 0.0

h_values = [0.01, 0.05, 0.1, 0.2, 0.5, 1.0]

plt.figure(figsize=(10, 6))

for h in h_values:
    # Coefficient matrix for Euler's Method
    A = np.array([[1, h],
                  [-k * h, 1]])

    # Eigenvalues
    eigvals = np.linalg.eigvals(A)

    print(f"Step size h = {h}")
    print("Coefficient matrix:")
    print(A)
    print("Eigenvalues:")
    print(eigvals)

    if np.all(np.abs(eigvals) < 1):
        print("Solution is Stable")
    else:
        print("Solution may not be Entirely Stable ;)")
    print()

    # Numerical solution
    x_values = np.arange(x_start, x_end + h, h)
    if x_values[-1] > x_end:
        x_values = x_values[:-1]

    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    y_values[0] = y0
    v_values[0] = v0

    for i in range(1, len(x_values)):
        y_values[i] = y_values[i - 1] + h * v_values[i - 1]
        v_values[i] = v_values[i - 1] - h * k * y_values[i - 1]

    plt.plot(x_values, y_values, label=f"h = {h}")

plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Second Order Stability Analysis with Euler's Method")
plt.legend()
plt.grid(True)
plt.show()
```
